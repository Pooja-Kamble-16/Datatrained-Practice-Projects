```python
df.head()
```

Now, all the column values are converted to ordinal and there are no categorical values left.

Here, we see that the feature "veil-type" is 0 and not contributing to the data. So we will remove it.

```python
df=df.drop(["veil-type"],axis=1)
```

### Creating X and Y Split

```python
x=df.drop("class",axis=1)
y=df["class"]
```

### Scaling

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
MMS=MinMaxScaler()
x_scaled=MMS.fit_transform(x)
x=pd.DataFrame(x_scaled,columns=x.columns)
x.head()
```

# Building the Model

### Creating train_test_split

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

# Finding the Best Random State

```python
from sklearn.tree import DecisionTreeClassifier
maxAcc=0
maxRS=0

for i in range(1,200):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=i)
    mod=DecisionTreeClassifier()
    mod.fit(x_train,y_train)
    pred=mod.predict(x_test)
    acc=accuracy_score(y_test,pred)
    if acc > maxAcc:
        maxAcc=acc
        maxRS=i
        
print("Best Accuracy is ",maxAcc," on Random State ",maxRS)
```

Using the best Random State to create new train_test_split.

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=maxRS)
```

```python
#Importing the packages to build models
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn import metrics
```

### 1.DecisionTreeClassifier

```python
DTC=DecisionTreeClassifier()
DTC.fit(x_train,y_train)
pred=DTC.predict(x_test)
acc=classification_report(y_test,pred)
print(acc)
```

### 2.RandomForestClassifier

```python
RFC=RandomForestClassifier()
RFC.fit(x_train,y_train)
pred=RFC.predict(x_test)
acc=classification_report(y_test,pred)
print(acc)
```

### 3.SVC

```python
SC=SVC()
SC.fit(x_train,y_train)
pred=SC.predict(x_test)
acc=classification_report(y_test,pred)
print(acc)
```

### 4.LogisticRegression

```python
LR=LogisticRegression()
LR.fit(x_train,y_train)
pred=LR.predict(x_test)
acc=classification_report(y_test,pred)
print(acc)
```

Here, we observe that LinearRegression, RandomForestClassifier and SupportVectorClassifier are giving the highest accuracy of 100%. But we all know, it may or may not be due to overfitting, we are not sure about this. So we cannot confirm it as the best model. To be sure about the best model, we will go for the Cross Validation Score.

### Cross Validation Score

```python
from sklearn.model_selection import cross_val_score
```

```python
print(cross_val_score(DTC,x,y,cv=5).mean())
```

```python
print(cross_val_score(RFC,x,y,cv=5).mean())
```

```python
print(cross_val_score(SC,x,y,cv=5).mean())
```

```python
print(cross_val_score(LR,x,y,cv=5).mean())
```

From the above cross validation score, we can conclude that DecisionTreeClassifier is our best model with the least difference.

### Hyper Parameter Tuning

Now, let's apply Hyper Paramter Tuning on the best model to improve the accuracy score.

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters={'criterion':['gini','entropy'],
           'splitter':['best','random'],
           'max_depth':[4,5,6,7,8],
           'max_features':['auto','sqrt','log2']}
```

```python
GCV=GridSearchCV(DecisionTreeClassifier(),parameters,cv=5)
GCV.fit(x_train,y_train)
```

```python
GCV.best_params_
```

```python
Final_mod=DecisionTreeClassifier(criterion='gini',max_depth=8,max_features='log2',splitter='best')
Final_mod.fit(x_train,y_train)
pred=Final_mod.predict(x_test)
acc=accuracy_score(y_test,pred)
print(acc)
```

After Hyper Parameter tuning, the improved score of the best model Linear Regression is 100%

### Plotting AUC_ROC curve

```python
metrics.plot_roc_curve(Final_mod, x_test, y_test)
```

The AUC score is 100%.
